```python
import pandas as pd
import requests
import lxml
from bs4 import BeautifulSoup
from string import ascii_uppercase as alphabet
```

```python
years = [1930,1934,1938,1942,1946,1950,1954,1958,1962,1966,1970,1974,1978,1982,1986,1990,1994,1998,2002,2006,2010,2014,2018,2022]
dict_football ={}
```

```python
def get_matches(year):
    web = f"https://en.wikipedia.org/wiki/{year}_FIFA_World_Cup"
    response = requests.get(web)
    content = response.text
    soup = BeautifulSoup(content, "lxml")
    matches = soup.find_all("div", class_= "footballbox")
    home = []
    score = []
    away = []

    for match in matches:
        home.append(match.find("th", class_ = "fhome").get_text())
        score.append(match.find("th", class_ = "fscore").get_text())
        away.append(match.find("th", class_ = "faway").get_text())

    dict_fb = {"home": home, "score": score, "away" : away}
    df = pd.DataFrame(dict_fb)
    df["year"] = year

    return df
```

```python
for year, letra in zip(years, alphabet):
    df = get_matches(year)
    dict_football[f"{year}"] = df
```
